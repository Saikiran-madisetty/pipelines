```python
prefixes = ['Mr.', 'Mr', 'Miss.', 'Miss', 'Mrs.', 'Mrs', 'Dr.', 'Dr']
 
def clean_name(name):
    parts = name.split()
    if len(parts) == 2:
        return ' '.join(parts)
    elif len(parts) == 4:
        return ' '.join(parts[1:-1])
    elif len(parts) == 3:
        if parts[0] in prefixes:
            return ' '.join(parts[1:])
        else:
            return ' '.join(parts[:-1])
    else:
        return name  
 
df['name'] = df['name'].apply(clean_name)

```
