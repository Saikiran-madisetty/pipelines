```python
import pandas as pd
df1=pd.read_csv(r"C:\Users\Saikiran\Downloads\cus_data.csv")
df1
```

```python
df2=pd.read_csv(r"C:\Users\Saikiran\Downloads\transaction_data.csv")
df2
```

```python
df3=pd.merge(df1,df2,on='customer_id',how='inner')
df3
```

```python
df3.info()
```

```python
df3.isnull().sum()
```

```python
df3['amount']=df3['amount'].fillna(df3['amount'].mean())
```

```python
gd=df3.groupby(['customer_id','phone','email'])['amount'].sum()
high_value=gd[gd>1000].sort_values(ascending=False)
high_value
```

```python
high_value.to_csv(r"C:\Users\Saikiran\Downloads\high_value_cust.csv")
```
